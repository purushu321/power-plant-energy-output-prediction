````python
df=pd.read_excel("../data/energy.xlsx")
df.head()
````

````python
df.info()
````

````python
df.shape
````

````python
X=df.drop(columns="Energy")
Y=df["Energy"]
````

````python
###Splitting the dataset into training set and testing set
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
````

**random_state** is a seed value that controls the randomness of dataset splitting.
Setting a fixed value ensures that the same training and testing sets are generated every time the code is executed, making results reproducible and easier to compare

````python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
````

**StandardScaler** is used to standardize features by transforming them to have mean 0 and standard deviation 1. We apply fit_transform() on the training data to learn scaling parameters and transform() on the test data to use the same parameters, preventing data leakage and ensuring fair model evaluation.

````python
from keras.models import Sequential
from keras.layers import Dense
````

 **Sequential**: Used to create a neural network by stacking layers sequentially.
- **Dense**: Creates a fully connected layer where every neuron is connected to all neurons in the next layer.
- **Dense Parameters**:
  - `units`: Number of neurons.
  - `activation`: Activation function (e.g., relu, sigmoid).

````python
def ann_model_regressor():
    model=Sequential()
    model.add(Dense(input_dim=4,units=6,activation='relu',kernel_initializer='uniform'))
    model.add(Dense(units=6,activation='relu',kernel_initializer='uniform'))
    model.add(Dense(units=1,activation='relu',kernel_initializer='uniform'))
    model.compile(optimizer='adam',loss='mean_squared_error')
    return model
````

### ANN Regressor Model

- `Sequential()` : Creates the ANN model.
- `Dense()` : Adds fully connected layers.
- `input_dim=4` : Input layer receives 4 features.
- `units=6` : Hidden layer contains 6 neurons.
- `activation='relu'` : Uses ReLU activation function.
- `kernel_initializer='uniform'` : Initializes weights uniformly.
- `optimizer='adam'` : Optimizes weights during training.
- `loss='mean_squared_error'` : Loss function used for regression problems.
- `return model` : Returns the constructed ANN model.

````python
ann=ann_model_regressor()
````

````python
ann.summary()
````

### ANN Summary

- First Hidden Layer:
  - 4 input features
  - 6 neurons
  - Parameters = (4 weights + 1 bias) Г— 6 = 30

- Second Hidden Layer:
  - 6 inputs
  - 6 neurons
  - Parameters = (6 weights + 1 bias) Г— 6 = 42

- Output Layer:
  - 6 inputs
  - 1 neuron
  - Parameters = 6 weights + 1 bias = 7

- Total Parameters:
  - 30 + 42 + 7 = 79

````python
X_train.shape
````

````python
###Training the ANN model on training set
ann.fit(X_train,Y_train, batch_size=32,epochs=100)
````

### Training the ANN Model

```python
ann.fit(X_train, Y_train, batch_size=32, epochs=100)
```

- **fit()** : Trains the ANN model on the training dataset.
- **X_train** : Input features used for training.
- **Y_train** : Target values corresponding to X_train.
- **batch_size=32** : The model processes 32 samples at a time before updating weights.
- **epochs=100** : The entire training dataset is passed through the network 100 times.

### Key Terms

- **Epoch** : One complete pass through the entire training dataset.
- **Batch** : A subset of training samples processed together.
- **Iteration** : One weight update after processing a batch.

### Example

Dataset Size = 1000 samples

Batch Size = 32

Iterations per Epoch вүҲ 1000 / 32 = 32

Epochs = 100

Total Weight Updates вүҲ 32 Г— 100 = 3200
```

````python
##Predicting the Result
ypred_train=ann.predict(X_train)
ypred_test=ann.predict(X_test)
````

### Predicting Results Using the Trained ANN

```python
ypred_train = ann.predict(X_train)
ypred_test = ann.predict(X_test)
```

- **predict()** : Uses the trained ANN model to generate predictions.
- **ypred_train** : Predictions for the training dataset.
- **ypred_test** : Predictions for the testing dataset.

### Purpose

- Compare `ypred_train` with `Y_train` to evaluate training performance.
- Compare `ypred_test` with `Y_test` to evaluate testing performance.
- Helps measure how well the model generalizes to unseen data.
```

````python
from sklearn.metrics import mean_squared_error
print("MSE for train data:", mean_squared_error(Y_train,ypred_train))
print("MSE for test data:", mean_squared_error(Y_test,ypred_test))
````

````python
from sklearn.metrics import r2_score
print("MSE for train data:", r2_score(Y_train,ypred_train))
print("MSE for test data:", r2_score(Y_test,ypred_test))
````
